```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
reader=data.pivot(index='userid',columns='booktitle',values='bookrating').reset_index(drop=True)
reader
```

```python
reader.fillna(0, inplace=True)
reader
```

```python
user_sim =1-pairwise_distances(reader.values,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.index=data.userid.unique()
user_sim_df.columns=data.userid.unique()
```
